# Timing the Cellular Automata Function
This file displays the algorithmic complexity of running our cellular automata simulation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in csv file
timing = pd.read_csv("timedsim.csv")
timing
```

```python
sns.lineplot(data=timing, x = "size", y = "time")
plt.ylabel("Time (nanoseconds)")
plt.xlabel("Size (nxn)")
plt.title("Runtime vs. CA size")
```
